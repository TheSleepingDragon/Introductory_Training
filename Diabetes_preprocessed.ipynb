```python
#create a dataframe with all training data except the target column (Glucose)
df2 = df1. dropna()
#If you want to drop one single column, you can:
#X = df2.drop(columns="Glucose")

#To select specific columns:
X = df2[["BloodPressure","SkinThickness","Insulin","Age","Glucose"]]
X
```

```python
# Separate target values
y = df2["Outcome"].values
# See the first 5 rows
y[0:5]
# Another Way of writing this code is:
# y = df2[["Outcome"]]
# See the first 5 rows
# y.head()
```

```python
from sklearn.model_selection import train_test_split
#split dataset into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=262)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7,metric="canberra")
knn.fit(X_train,y_train)

#check accuracy of our model on the test data
knn.score(X_test, y_test)
```

```python
from sklearn import svm
#Create a svm Classifier
clf = svm.SVC(kernel='rbf', degree = 5, gamma = 0.001, C = 10) # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a svm Classifier#
clf=RandomForestClassifier(n_estimators=1000, criterion="entropy")

# Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

# Predict the response for test dataset
y_pred=clf.predict(X_test)
clf.score(X_test, y_test)
```

```python
from sklearn.neural_network import MLPRegressor

MLP = MLPRegressor(learning_rate_init=0.01, max_iter=100, momentum=0.9, hidden_layer_sizes=(10,10,10,10), random_state=262, activation="relu", learning_rate="adaptive").fit(X_train,y_train)
y_pred=MLP.predict(X_test)
MLP.score(X_test, y_test)
```
