## Importing Modules

```python
import matplotlib as mpl
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

## Simple Graph

```python
x = np.linspace(-5, 3, 30) #Lower number for the last arguement, may result in segmented plots (try it!).
```

```python
y1 = x**4 - 5*x**2 - 10*x - 2
```

```python
y2 = 4*x**3 - 10*x - 10
```

```python
y3 = 12*x**2 - 10
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, color="peru", label="y(x)")
ax.plot(x, y2, color="steelblue", label="y'(x)")
ax.plot(x, y3, color="purple", label="y''(x)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```

## Figure

```python
fig = plt.figure(figsize=(8, 3), facecolor="beige")

#axes
left, bottom, width, height = 1, 1, 1, 1
ax = fig.add_axes((left, bottom, width, height), facecolor="ivory")

x = np.linspace(-2, 2, 2000)
y1 = np.cos(10 * x)
y2 = np.exp(-x**2)

ax.plot(x, y1 * y2)
ax.plot(x, y2, 'sienna')
ax.plot(x, -y2, 'chocolate')
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.savefig("Functions.png", dpi=150)
```
