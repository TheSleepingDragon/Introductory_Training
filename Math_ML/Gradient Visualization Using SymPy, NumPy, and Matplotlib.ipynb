```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import sympy
%matplotlib inline
```

## Generating a 2D Meshgrid

```python
nx, ny = (10,10)

x, y = sympy.symbols('x y')
f = sympy.Function('f')(x, y)
expr = 2*x - (y**2)

x1 = np.linspace(-5, 10, nx)
y1 = np.linspace(-5, 10, ny)

xg, yg = np.meshgrid(x1, y1)
```

```python
xg
```

```python
f = sympy.lambdify((x, y), expr, 'numpy') # "Numpy in sympy" for computation is faster than in-built sympy tools.
```

```python
# z is the product of xg and yg 
z = f(xg, yg)
z
```

### Making a color plot to display the data

```python
plt.figure(figsize=(12,12))
plt.pcolor(xg, yg, z)
plt.title('Color Plot of f(x, y) = x * y^2')
plt.colorbar()
```

```python
Gy, Gx = np.gradient(z) #ATTENTION! NOTE HOW X AND Y ARE ORDERED IN THIS LINE.
```

```python
Gy
```

```python
Gx
```

```python
plt.figure(figsize=(12,12))
plt.pcolor(xg, yg, z)
plt.title('Gradient Plot of f(x, y) = x * y^2')
plt.colorbar()
plt.quiver(xg, yg, Gx, Gy, scale=500, color='w')
```

### Calculate the partial derivatives for the gradient

```python
ddx = sympy.diff(f(x, y), x)
ddy = sympy.diff(f(x, y), y)
```

```python
print(ddx, ddy, sep='\n')
```

```python
# Without sympy, the codes should be like this:
# def ddx(x, y):
#     return sympy.diff(f(x, y), x)
# def ddy(x, y):
#     return sympy.diff(f(x, y), y)
```

```python
Gx = sympy.lambdify(y, ddx, 'numpy')
Gy = sympy.lambdify((x, y), ddy, 'numpy')
```

```python
plt.figure(figsize=(12,12))
plt.pcolor(xg, yg, z)
plt.title('Plot of y^2, 2xy')
plt.colorbar()
A = Gx(yg)
B = Gy(xg, yg)
plt.quiver(xg, yg, A, B, scale=500, color='w')
```

### The figure indicates that x values close to 10 and y values close to 0 generate the largest values!
